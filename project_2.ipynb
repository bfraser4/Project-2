```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract CSVs into DataFrames
disney_file = "Resources/disney_plus_titles.csv"
disney_df = pd.read_csv(disney_file)
disney_df.head()
```

```python
# extract CSV 2 into df
netflix_file = "Resources/netflix_titles.csv"
netflix_df = pd.read_csv(netflix_file)
netflix_df.head()
```

```python
# extract CSV 3 into df
imdb_file = "Resources/Top 1000 IMDB movies.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df.head()
```

```python
imdb_cols = ["Movie Name", "Movie Rating", "Votes"]
imbd_transformed= imdb_df[imdb_cols].copy()
```

```python
# #Disney+ Create a  filtered dataframe from specific columns
# imdb_cols = ["Movie Name", "Movie Rating", "Votes"]
# imbd_transformed= imdb_df[imdb_cols].copy()

# Rename the column headers
imdb_transformed_title = imbd_transformed.rename(columns={"Movie Name": "title"})

# Clean the data by dropping duplicates and setting the index
#disney_transformed.drop["show_id"] inplace=True)
#disney_transformed.set_index("show_id", inplace=True)

imdb_transformed_title.head()
```

```python
#Disney+ Create a  filtered dataframe from specific columns
disney_cols = ["title", "date_added", "rating"]
disney_transformed= disney_df[disney_cols].copy()

# Rename the column headers
#premise_transformed = disney_transformed.rename(columns={"License Serial Number": "id",
#                                                          "Premises Name": "premise_name",
#                                                          "County ID Code": "county_id"})

# Clean the data by dropping duplicates and setting the index
#disney_transformed.drop["show_id"] inplace=True)
#disney_transformed.set_index("show_id", inplace=True)

disney_transformed.head()
```

```python
#netflix Create a  filtered dataframe from specific columns
netflix_cols = ["title", "date_added", "rating"]
netflix_transformed= netflix_df[netflix_cols].copy()

# Rename the column headers
# netflix_transformed = netflix_transformed.rename(columns={"License Serial Number": "id",
#                                                           "Premises Name": "premise_name",
#                                                           "County ID Code": "county_id"})

# Clean the data by dropping duplicates and setting the index
#netflix_transformed.drop_duplicates("id", inplace=True)
#netflix_transformed.set_index("id", inplace=True)

netflix_transformed.head()
```

```python
stream_db= pd.merge(left=netflix_transformed,right=disney_transformed, how = 'left', on = "title")
stream_db.head(100)
stream_db = stream_db.drop(columns={"date_added_y", "rating_y"})
stream_db= stream_db.rename(columns={"date_added_x": "date_added", "rating_x": "rating"})
stream_db
```

```python
imdb_rating_merge= pd.merge(left=stream_db,right=imdb_transformed_title, how = 'left', on = 'title')
imdb_rating_merge.dtypes
```

```python

#create database conn

connection_string = "postgres:postgres@localhost:5432/streaming_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#database connection string
protocol = 'postgresql'
username = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432
database_name = 'streaming_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
#append to streaming_db
imdb_rating_merge.to_sql(name='streaming_db', con=engine, if_exists='append', index=False)
```
